```python
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
ticker_sym = ["NFLX", "AMD"]
data = yf.download(ticker_sym,
                   start='2023-02-01',
                   end='2024-01-01',
                   progress=False)# downloads historical stock data for the specified ticker symbols.
data.head()
```

```python
col = ["Adj Close"]
data.drop(columns=col , axis =1, inplace = True)#Removing the unwanted columns in the dataset
data.head()
```

```python
data.describe()
```

```python
closing_prices = data['Close']
closing_prices
```

```python
Max  = closing_prices.max()
Min = closing_prices.min()
print(f'Max closing price for both companies:\n{Max}')
print(f'Min closing price for both companies:\n{Min}')
```

```python
closing_prices.plot(figsize=(10, 6))
plt.title('Stock Performance Comparison of both companies')
plt.ylabel('Closing Price (USD)')
plt.grid(True)
plt.show()
```

***We can see that NFLX trades at a higher stock price level compared to AMD throughout most of the time period shown.<br>
AMD's stock price experiences more frequent and sharper ups and downs compared to NFLX.<br>
We can also see that towards the right side of the plot both stocks gets an upward trend with NFLX experiencing a steeper rise in its stock price.***



```python
price_changes = closing_prices.pct_change() * 100
```

```python
price_changes.plot(figsize=(10, 6))
plt.title('Percentage change in Stock Prices')
plt.grid(True)
plt.show()
```

```python
volatility = closing_prices.pct_change().std() * (252 ** 0.5)
avg_vol = data['Volume'].mean()
returns = closing_prices.pct_change().mean() * 252
print(f'Volatility:\n{volatility}\n')
print(f'Average Trading Volume:\n{avg_vol}\n')
print(f'Return on Investment:\n{returns}\n')
```

```python
volume_traded = data["Volume"]
volume_changes = volume_traded.pct_change() * 100
print(f' %Volume changes:\n{volume_changes}\n')
```

```python
volume_changes.plot(figsize=(10, 6))
plt.title('Percentage Changes in Volume Traded')
plt.grid(True)
plt.show()
```

### This plot displays the percentage changes in trading volume over the time for two stocks represented by their symbols.
***The y-axis shows the percentage change in volume, ranging from 0% to around 400% and the x-axis represents the time period.<br>
The orange bars represent the volume fluctuations for NFLX stock, while the blue bars depict the volume changes for AMD stock.
The volatility seems to be higher for NFLX compared to AMD as indicated by the  orange bars.***

```python
moving_avg_20 = closing_prices.rolling(window=20).mean()
moving_avg_50 = closing_prices.rolling(window=50).mean()
moving_avg_200 = closing_prices.rolling(window=200).mean()
```

```python
ma20_changes = moving_avg_20.pct_change() * 100
ma50_changes = moving_avg_50.pct_change() * 100
ma200_changes = moving_avg_200.pct_change() * 100
```

```python
plt.figure(figsize=(12, 8))
ma20_changes.plot(label='20-day MA')
plt.title('Percentage Changes in Moving Averages for 20-day')
plt.show()
```

***We can observe that :<br>
Around mid-2023 both stocks experienced a  spike in their 20-day moving average percentage changes with AMD reaching a higher peak.<br>
Towards the end of the time period both stocks exhibit an upward trend with NFLX having a more  positive percentage change in its 20-day moving average.***

```python
plt.figure(figsize=(12, 8))
ma50_changes.plot(label='50-day MA',linestyle= ':')
plt.title('Percentage Changes in Moving Averages for 50-day')
plt.show()
```

***Around September-October 2023 both stocks experienced a notable dip in their 50-day moving average percentage changes.<br>
Towards the end of the time period  NFLX had positive trend in its 50-day moving average percentage change compared to AMD.***

```python
plt.figure(figsize=(12, 8))
ma200_changes.plot( label='200-day MA', linestyle='--')
plt.title('Percentage Changes in Moving Averages for 200-day')
```

***We can observe that  NFLX seems to have a more consistent upward trajectory in its 200-day moving average percentage change compared to AMD.***
